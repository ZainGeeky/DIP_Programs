```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a signal
t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 10 * t)

# Rotation angle in radians
rotation_angle = np.pi/4

# Perform time domain rotation
rotated_signal = np.cos(rotation_angle) * signal + np.sin(rotation_angle) * np.roll(signal, -int(len(signal) * rotation_angle / (2 * np.pi)))

# Plot the original and rotated signals
plt.plot(t, signal, label='Original')
plt.plot(t, rotated_signal, label='Rotated')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
